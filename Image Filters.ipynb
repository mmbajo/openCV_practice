```python
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt
from PIL import Image, ImageFilter
%matplotlib inline
```

```python
image = cv2.imread('sample.bmp')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (11, 6))
plt.imshow(image)
plt.title('Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
image = cv2.imread('sample.bmp')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
figure_size = 9 # the dimension x, y of the kernel (filter)
new_image = cv2.blur(image, (figure_size, figure_size))

plt.figure(figsize = (11, 6))
plt.subplot(121)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_HSV2RGB))
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_HSV2RGB))
plt.title('Mean Filter')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
image2 = cv2.cvtColor(image, cv2.COLOR_HSV2BGR)
image2 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
figure_size = 9 # the dimension x, y of the kernel (filter)
new_image = cv2.blur(image2, (figure_size, figure_size))

plt.figure(figsize = (11, 6))
plt.subplot(121)
plt.imshow(image2, cmap = 'gray')
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(new_image, cmap = 'gray')
plt.title('Mean Filter')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
new_image = cv2.medianBlur(image, figure_size)

plt.figure(figsize = (11, 6))
plt.subplot(121)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_HSV2RGB))
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_HSV2RGB))
plt.title('Median Filter')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
new_image = cv2.medianBlur(image2, figure_size)

plt.figure(figsize = (11, 6))
plt.subplot(121)
plt.imshow(image2, cmap = 'gray')
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(new_image, cmap = 'gray')
plt.title('Median Filter')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
new_image = cv2.GaussianBlur(image, (figure_size, figure_size), 0)

plt.figure(figsize = (11, 6))
plt.subplot(121)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_HSV2RGB))
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_HSV2RGB))
plt.title('Gaussian Filter')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
new_image = cv2.Canny(image, 35, 70)

plt.figure(figsize = (11, 6))
plt.subplot(121)
plt.imshow(image)
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(new_image)
plt.title('Canny Edge Detection')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
a = 1
kernel = np.array([[a, 0, a],
                   [a, 0, a],
                   [a, 0, a]])

new_image = cv2.filter2D(image, -1, kernel)
plt.figure(figsize = (11, 6))
plt.subplot(121)
plt.imshow(image)
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(new_image)
plt.title('Customized Filter: a = {}'.format(a))
plt.xticks([])
plt.yticks([])
plt.show()
```
